```python
import matplotlib.pyplot as plt
import numpy as np
x =  np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.exp(x), '-g', label = 'sin(x)')
plt.plot(x, np.log(x), ':b', label = 'cos(x)')
plt.xlabel('x axis')
plt.legend()

```

```python
x= np.random.randn(1000) #mu=0, sigma=1
plt.hist(x) #uni-variate Gaussian
```

```python
plt.hist(x, bins=30, density=True, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='none')
```

```python
#multivariate Gaussian and 2-D histograms
mean = [0,0]
cov = [[1, 1], [1, 2]] #covariance matrix
x, y = np.random.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x, y, bins=30, cmap='Blues')
cb= plt.colorbar()
cb.set_label('counts in bin')
```
